```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normalize_image(image_path,R=128):
    img = cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)
    img_float = img.astype(np.float32)
    min_val, max_val = np.min(img_float), np.max(img_float)
    norm_img = np.round((img_float - min_val)/(max_val-min_val) * R).astype(np.uint8)
    plt.subplot(1,2,1)
    plt.title("normalized Image")
    plt.imshow(norm_img,cmap='gray')
    plt.subplot(1,2,2)
    plt.title("Original image")
    plt.imshow(img,cmap='gray')

image_path = "monkey.jpg"
normalize_image(image_path)
```

```python
def negative_image(image_path,L=128):
    img=cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)
    #img_float=img.astype(np.float32)
    negative_img =(L-1-img)
    plt.imshow(negative_img,cmap='gray')

image_path = "monkey.jpg"
negative_image(image_path)
```

```python
def log_image(image_path,L=128):
    img=cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)
    #img_float=img.astype(np.float32)
    log_img = (1*np.log(1+img))
    plt.imshow(log_img,cmap='gray')

image_path = "monkey.jpg"
negative_image(image_path)
```
