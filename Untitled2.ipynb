```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def image_negative(img):
    return 255 - img
def identity_transformation(img):
    return img.copy()
def log_transformation(img):
    c = 255 / np.log(1 + np.max(img))
    return (c * np.log(1 + img)).astype(np.uint8)
def inverse_log_transformation(img):
    c = 255 / (np.exp(np.max(img)) - 1)
    return (c * (np.exp(img) - 1)).astype(np.uint8)
def power_law_transformation(img, gamma):
    c = 255 / (np.max(img) ** gamma)
    return (c * (img ** gamma)).astype(np.uint8)
def nth_root_transformation(img, n):
    return (255 * (img / 255) ** (1 / n)).astype(np.uint8)
def nth_power_transformation(img, n):
    return (255 * (img / 255) ** n).astype(np.uint8)
def apply_transformations(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    
    transformations = {
        "Original": img,
        "Negative": image_negative(img),
        "Identity": identity_transformation(img),
        "Log": log_transformation(img),
        "Inverse Log": inverse_log_transformation(img),
        "Power Law (Gamma=0.5)": power_law_transformation(img, 0.5),
        "Power Law (Gamma=2)": power_law_transformation(img, 2),
        "Nth Root (n=3)": nth_root_transformation(img, 3),
        "Nth Power (n=3)": nth_power_transformation(img, 3),
    }
    
    plt.figure(figsize=(12, 8))
    for i, (title, transformed_img) in enumerate(transformations.items()):
        plt.subplot(3, 3, i+1)
        plt.imshow(transformed_img, cmap='gray')
        plt.title(title)
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

image_path = "images.jpg"  # Provide the path to the image
apply_transformations(image_path)
```

```python
def bit_plane_slicing(image_path):
    # Load the grayscale image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    if image is None:
        print("Error: Image not found!")
        return

    # Get image dimensions
    M, N = image.shape

    # Create a figure to visualize
    fig, axes = plt.subplots(2, 4, figsize=(12, 6))
    axes = axes.ravel()

    # Display original image
    axes[0].imshow(image, cmap='gray')
    axes[0].set_title("Original Image")
    axes[0].axis('off')

    # Extract and display each bit plane
    for i in range(8):
        bit_plane = (image >> i) & 1  # Extract bit i
        bit_plane = bit_plane * 255   # Scale for visualization
        
        axes[i+1].imshow(bit_plane, cmap='gray')
        axes[i+1].set_title(f"Bit Plane {i}")
        axes[i+1].axis('off')

    # Show the figure
    plt.tight_layout()
    plt.show()

# Example Usage
bit_plane_slicing("images.jpg")
```

```python
image = cv2.imread("images.jpg", cv2.IMREAD_GRAYSCALE)
M, N = image.shape

# Generate a random binary message of size M*N
message = np.random.randint(0, 2, (M, N), dtype=np.uint8)

# Embed message into the LSB of the image
embedded_image = image.copy()
embedded_image = (embedded_image & 0xFE) | message  # Modify LSB

# Extract the message from the embedded image
extracted_message = embedded_image & 1

# Plot original and modified images
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].imshow(image, cmap='gray')
axes[0, 0].set_title("Original Image")
axes[0, 1].imshow(embedded_image, cmap='gray')
axes[0, 1].set_title("Embedded Image")
axes[1, 0].imshow(extracted_message, cmap='gray')
axes[1, 0].set_title("Extracted Message")

# Plot histograms
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.hist(image.ravel(), bins=256, color='blue', alpha=0.7)
plt.title("Histogram of Original Image")

plt.subplot(1, 3, 2)
plt.hist(embedded_image.ravel(), bins=256, color='red', alpha=0.7)
plt.title("Histogram of Embedded Image")

plt.subplot(1, 3, 3)
plt.hist(extracted_message.ravel(), bins=256, color='green', alpha=0.7)
plt.title("Histogram of Extracted Message")

plt.show()
```

```python
image = cv2.imread("images.jpg", cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Plot the original and equalized images
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(2, 2, 1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis("off")

# Equalized Image
plt.subplot(2, 2, 2)
plt.imshow(equalized_image, cmap="gray")
plt.title("Equalized Image")
plt.axis("off")

# Histogram of Original Image
plt.subplot(2, 2, 3)
plt.hist(image.ravel(), bins=256, range=[0, 256], color="black")
plt.title("Histogram of Original Image")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")

# Histogram of Equalized Image
plt.subplot(2, 2, 4)
plt.hist(equalized_image.ravel(), bins=256, range=[0, 256], color="black")
plt.title("Histogram of Equalized Image")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")

# Show the plots
plt.tight_layout()
plt.show()
```
