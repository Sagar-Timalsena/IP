```python
from PIL import Image
from matplotlib import pyplot as plt
img = Image.open("CLena.gif")
```

# Converting it into Gray Scale

```python
gray_img = img.convert("L")
```

# Resize it

```python
small_img64 = img.resize((64,64))
small_img128 = img.resize((128,128))
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(4,4))
ax=axes.ravel()
ax[0].imshow(img,cmap=plt.cm.gray)
ax[0].set_title('original Image')
ax[1].imshow(gray_img,cmap=plt.cm.gray)
ax[1].set_title('gray Image')
ax[2].imshow(small_img64,cmap=plt.cm.gray)
ax[2].set_title('small image64')
ax[3].imshow(small_img128,cmap=plt.cm.gray)
ax[3].set_title('small image128')
for a in ax:
    a.axis('off')
plt.tight_layout()
plt.show()
```
