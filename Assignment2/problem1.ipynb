```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the color image
image = cv2.imread("../test_imgs/Lenna.jpg")  # Replace with the actual image path

# Check if the image is loaded properly
if image is None:
    raise FileNotFoundError("Image not found. Please check the file path.")
    
# Print size and data type
print(f"Image Shape: {image.shape}")  # (Height, Width, Channels)
print(f"Data Type: {image.dtype}")

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imwrite("grayscale_image.jpg", gray_image)  # Save the grayscale image

# Display the grayscale image as an array
print("Grayscale Image Array:")
print(gray_image)

# Print the first row of the matrix
first_row = gray_image[0].tolist()
print("First Row of Grayscale Image:")
print(first_row)

# Print a sub-region (top-left 10x10 pixels)
sub_region = gray_image[:10, :10]
print("Top-left 10x10 Pixels:")
print(sub_region)

# Plot original and grayscale images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct display
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")

plt.savefig("output_imgs/problem1.png")  
plt.show()
```
