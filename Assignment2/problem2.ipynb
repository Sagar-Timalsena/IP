```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def resize_images(img1, img2):
    """Resize images to the same size if they are different."""
    if img1.shape[:2] != img2.shape[:2]: 
        img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))
        print("Images were different sizes. Resized img2 to match img1.")
    else:
        print("Both images are already the same size.")
    return img1, img2

def perform_operations(img1, img2):
    """Perform arithmetic operations on two images."""
    addition = cv2.add(img1, img2)
    subtraction = cv2.subtract(img1, img2)
    multiplication = cv2.multiply(img1, img2)
    averaging = (img1 // 2) + (img2 // 2)  
    return addition, subtraction, multiplication, averaging

# Read the images
img1 = cv2.imread("../test_imgs/Lenna.jpg", cv2.IMREAD_UNCHANGED)
img2 = cv2.imread("../test_imgs/cameraman.jpg", cv2.IMREAD_UNCHANGED)

if img1 is None or img2 is None:
    raise FileNotFoundError("One or both images not found. Please check file paths.")

# Resize 
img1, img2 = resize_images(img1, img2)

# Perform arithmetic operations and store the results in these variables
add_img, sub_img, mul_img, avg_img = perform_operations(img1, img2)

# Plot results
titles = ["Image 1", "Image 2", "Addition", "Subtraction", "Multiplication", "Averaging"]
images = [img1, img2, add_img, sub_img, mul_img, avg_img]

plt.figure(figsize=(10, 10))
for i in range(len(titles)):
    plt.subplot(3, 2, i + 1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis("off")

plt.savefig("output_imgs/problem2.png") 
plt.show()
```
