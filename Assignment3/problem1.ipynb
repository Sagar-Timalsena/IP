# Least Significant Bit (LSB) Modification in an Image

    * Load an Image: Choose a grayscale.
    * Modify the Image: Convert the image into its pixel matrix.
    * Set the Least Significant Bit (LSB) of each pixel to zero.
    * Save the modified image.
    * Display the Images: Display the original image and Display the modified image.
    * Compute and display the difference image (original - modified).
    * Analyze the Difference:  the visual and numerical differences between the images.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def modify_LSB(img):
    """ This function modify the image.
        Args: img """
    # convert image into its pixel matrix using numpy array
    img_arr = np.array(img, dtype = np.uint8)
    
    # set the LSB of each pixel to zero
    modified_img = img_arr & 0xFE
    return modified_img

def main():
    # read grayscape image
    org_img = cv2.imread(f"../test_imgs/lena.jpg", cv2.IMREAD_GRAYSCALE)
    if org_img is None:
        print("Error: image not found.")
        return

    modf_img = modify_LSB(org_img)
    diff_img = org_img - modf_img  # difference of the image
    
    # Display the images
    plt.figure(figsize=(8, 4))
    
    plt.subplot(1, 3, 1)
    plt.imshow(org_img, cmap='gray')
    plt.title("Original Image")
    plt.axis("off")
    
    plt.subplot(1, 3, 2)
    plt.imshow(modf_img, cmap='gray')
    plt.title("Modified Image")
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.imshow(diff_img, cmap='gray')
    plt.title("Difference Image")
    plt.axis("off")
    
    plt.savefig("outimg_problem1.png")
    plt.show()
    
    # Analyze the numerical differences
    print("Numerical Analysis:")
    print(f"Min Difference: {np.min(diff_img)}")
    print(f"Max Difference: {np.max(diff_img)}")
    print(f"Mean Difference: {np.mean(diff_img)}")

if __name__ == "__main__":
    main()
```
