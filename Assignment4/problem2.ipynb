# Implement a function to perform bit plane slicing on the 8-bit grayscale image of size M*N.

    * The function should extract individual bit planes ranging from 0 to 7.
    * Visualize the original grayscale image along with its individual bit planes.

```python
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt

def read_image(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError(f"Error: Unable to read image at {image_path}")
    return img

def bit_plane_slicing(image):
    bit_planes = [(image >> i) & 1 for i in range(8)]
    return bit_planes

def plot_bit_planes(original, bit_planes):
    plt.figure(figsize=(12, 8))
    plt.subplot(3, 3, 1)
    plt.imshow(original, cmap='gray')
    plt.title("Original Image")
    plt.axis("off")

    for i, plane in enumerate(bit_planes, start=2):
        plt.subplot(3, 3, i)
        plt.imshow(plane * 255, cmap='gray')  # Scale to 255 for visibility
        plt.title(f"Bit Plane {i-2}")
        plt.axis("off")

    plt.savefig("outimg_problem2.png")
    plt.show()

def main():
    img_path = "../test_imgs/lena.jpg"  # Update with your image path
    img = read_image(img_path)

    bit_planes = bit_plane_slicing(img)
    plot_bit_planes(img, bit_planes)

if __name__ == "__main__":
    main()
```
